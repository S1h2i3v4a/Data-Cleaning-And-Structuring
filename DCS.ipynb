```python
import json
```

```python
def load_data(file_path):
    """Load JSON data from file, normalize types, and remove duplicate names."""
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
    clean_data = []
    unique_names = set()
    for user in data:
        # Safe defaults and normalization
        name = user.get('name') if user.get('name') is not None else ''
        age = user.get('age') if 'age' in user else None
        rating = user.get('rating') if 'rating' in user else None
        # Normalize rating to int when it's a digit string
        try:
            if isinstance(rating, str) and rating.isdigit():
                user['rating'] = int(rating)
        except Exception:
            pass
        # Ensure age key exists
        if age is None:
            user['age'] = None
        # Skip duplicates by name
        if name not in unique_names:
            unique_names.add(name)
            clean_data.append(user)
    return clean_data
```

```python
data = load_data('store_data.json')
print(data)
```
